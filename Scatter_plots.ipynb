```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.scatter(y,x, c='r', label='Chicago')
plt.title('PyBer Fare by Month')
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.legend()
plt.gca().invert_yaxis()
plt.xlim(0,50)
```

```python
plt.scatter(x,y)

y_larger = []
for data in y: 
    y_larger.append(data * 3)
```

```python
plt.scatter(x,y, s=y_larger)
```

```python
# List comprehension way to use python for For loops built in
# new_list = [expression for item in list if conditional]

plt.scatter(x,y, s = [i*3 for i in y])
```

```python
#OOI method for making scatter plots

fig, ax = plt.subplots()
ax.scatter(x,y)
```

```python
#OOI method for making bubble charts

fig, ax = plt.subplots()
ax.scatter(y,x, s= [i*5 for i in y], c= 'skyblue', alpha = .8, edgecolor='k', label = 'Boston')
ax.legend()
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
ax.invert_yaxis()
ax.set_xlim(0,50)
```
